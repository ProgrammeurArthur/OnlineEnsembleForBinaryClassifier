```python
# Função para carregar os dados
def load_data():
    all_logs = {}
    for dataset_name in path_results:
        all_logs[dataset_name] = []
        for ensemble, simulations in ensemble_simulations.items():
            for simulation in simulations:
                df = pd.read_csv('{}/{}/{}.csv'.format(dataset_name, ensemble, simulation), sep=',')
                # Obter apenas a última linha de cada arquivo CSV
                df = df.iloc[[-1]]
                # Extrair apenas a métrica "Accuracy"
                accuracy_value = df.iloc[0]["Accuracy"]
                # Adicionar o valor de acurácia à lista
                all_logs[dataset_name].append(accuracy_value)
    return all_logs

# Carregar os dados
all_logs = load_data()

# Criar uma tabela PrettyTable
table = p.PrettyTable()

# Adicionar os cabeçalhos à tabela
table.field_names = ["Algorithm"] + path_results

# Preencher a tabela com os dados de acurácia
for ensemble, simulations in ensemble_simulations.items():
    for simulation in simulations:
        accuracy_values = []
        for dataset_name in path_results:
            # Remover o sufixo "_resultados" dos nomes dos algoritmos
            algorithm_name = simulation.replace('_resultados', '')
            # Encontrar o valor de acurácia para o último conjunto de dados
            accuracy = all_logs[dataset_name][simulations.index(simulation)]
            accuracy_values.append(accuracy)
        # Adicionar os valores de acurácia à linha da tabela
        table.add_row([algorithm_name] + accuracy_values)

# Imprimir a tabela
print(table)
```
